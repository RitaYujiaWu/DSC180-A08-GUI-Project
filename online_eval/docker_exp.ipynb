```python
from IPython.display import Image, display
import logging

logging.basicConfig(level=logging.INFO, force=True)

from openrlhf.env.android_world_env import RemoteAndroidWorldEnv

env = RemoteAndroidWorldEnv(
    base_url="http://localhost",
    env_port=5003,
)

try:
    observation = env.reset()
    if observation.get("screenshot"):
        display(Image(data=observation["screenshot"], width=320))
    else:
        print("No screenshot on reset.")

    action_payload = {"action_type": "open_app", "app_name": "Chrome"}
    obs, reward, done, info = env.step(action_payload)
    print(reward, done, info)
    if obs.get("screenshot"):
        display(Image(data=obs["screenshot"], width=320))
    else:
        print("No screenshot after step.")

finally:
    env.close()
```
