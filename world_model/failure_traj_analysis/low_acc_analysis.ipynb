# **Failure-Aware Continuous Memory: Detailed Analysis of Successful vs. Failed Trajectories**

## **Background & Motivation**
In our previous work, we demonstrated that GUI agents equipped with **continuous memory** can learn effectively from successful trajectories. These embeddings preserve rich visual and action-level semantics while maintaining fixed computational cost. However, only ~20% of collected trajectories are successful, meaning **80% of our data is unused**.  
This unused portion contains *highly valuable failure signals* that reveal:
- where the agent becomes confused,
- what UI states cause oscillation or resets,
- how interruptions affect progress,
- and what mistakes commonly prevent task completion.

Learning from failures is essential because:
1. **Failure patterns are consistent across GUIs** (e.g., repeated misclick loops, context resets, premature terminations).
2. **Successful trajectories alone do not teach avoidance behaviors**.
3. **Continuous memory embeddings naturally support contrastive learning**, making it ideal for success vs. failure differentiation.

Our goal is to build a **self-evolving GUI agent** that uses continuous memory to:
- detect error trajectories,
- extract failure signatures,
- connect failures to corrected outcomes,
- and improve policy robustness at inference time without retraining.

The following detailed analysis shows how we can extract structured insights by comparing a pair of successful and failed trajectories on *Allrecipes*, illustrating concrete failure patterns and how to leverage them in continuous memory.

---

# **1. High-Level Comparison: Why Success Succeeded / Why Failure Failed**

## ‚úÖ Successful Trajectory

1. Search for ‚Äúbanana bread‚Äù.

### **Key Behavioral Attributes**

# **2. Extracted Failure Signatures (What the Agent is Doing Wrong)**

This section converts trajectory differences into **structured, reusable failure patterns** that can be embedded in continuous memory.

---

## **Failure Pattern 1 ‚Äî Context Reset Loop**
**Symptoms:**  
- Returning to homepage after already being on correct results page.  
- Re-typing ‚ÄúGreek salad‚Äù even though the search results are already visible.

**Underlying cause:**  
The agent misidentifies its UI context, failing to recognize that it is already in the correct state.

**Memory abstraction:**  
- Create continuous tokens representing ‚Äúalready in search results ‚Üí do not search again‚Äù.
- Penalize repeated search embeddings via contrastive learning.

---

## **Failure Pattern 2 ‚Äî Filter Oscillation**
**Symptoms:**  
- Clicking the filter icon multiple times with no actual filtering applied.
- Entering and exiting filter menus repeatedly.

**Underlying cause:**  
The agent recognizes filter affordances but does not connect them to a multi-step workflow that requires commitment.

**Memory abstraction:**  
- Store successful filter usage sequences.
- Store oscillation failure embeddings as negative prototypes.
- During inference, if current embedding matches an oscillation pattern, increase the likelihood of *completing* the next filter-related step.

---

## **Failure Pattern 3 ‚Äî Inability to Transition to Recipe Detail Page**
**Symptoms:**  
- Repeatedly scrolling or interacting with filters but never opening a recipe.
- Ignoring recipe cards even when visible.

**Underlying cause:**  
The agent lacks a strong prior for transitioning from results ‚Üí detail view.

**Memory abstraction:**  
- Encode successful transitions as canonical workflow fragments.
- Encode ‚Äústaying on results page for too long‚Äù as a failure signature.
- Retrieval steers the agent toward detail-page-opening actions.

---

## **Failure Pattern 4 ‚Äî Poor Popup / Error Recovery**
**Symptoms:**  
- ‚ÄúPlease fill this field‚Äù popup interrupts agent, causing a context reset loop.

**Underlying cause:**  
The agent does not generalize recovery strategies from unexpected UI events.

**Memory abstraction:**  
- Create popup-recovery embeddings (negative + positive guidance pairs).
- Retrieval during inference helps suppress search-reset behavior.

---

## **Failure Pattern 5 ‚Äî Premature Task Termination**
**Symptoms:**  
- Ends with ‚ÄúEarly stop: Reach max steps 15‚Äù without verifying any recipe.

**Underlying cause:**  
The agent wastes too many steps in loops, leaving insufficient budget for task completion.

**Memory abstraction:**  
- Failure token representing ‚Äúno verification performed after extended navigation‚Äù.
- Inference-time penalty when the agent approaches such states.

---

# **3. Root Cause Analysis: Why the Failure Patterns Occur**

From the failed trajectory, we can deduce these deeper systemic issues:

---

## **Root Cause A ‚Äî Misclassification of UI States**
The agent treats:
- search results page as if it is the homepage,
- filter menu as if it requires reactivation,
- popups as terminal states.

This leads to repeated resets and incorrect actions.

### Memory Solution
Store prototype embeddings of:
- search result screens  
- filter screens  
- detail pages  
- popup/error states  

Use similarity search to stabilize UI state recognition.

---

## **Root Cause B ‚Äî Weak Multi-Step Task Planning**
Successful filtering requires:
1. entering filter menu,  
2. selecting criteria,  
3. applying filters.  

The failed trajectory only accomplishes (1) repeatedly.

### Memory Solution
Store ‚Äúfilter workflow embeddings‚Äù extracted from successful trajectories:
- FilterEntry ‚Üí FilterSelection ‚Üí FilterApply ‚Üí RecipeSelect

Retrieve these embeddings during inference to guide the agent.

---

## **Root Cause C ‚Äî No Recipe Verification Habit**
The agent never reaches or validates:
- prep time,  
- review count,  
- cheese type,  
- dressing type.  

It lacks the final verification reflex.

### Memory Solution
Store ‚Äúverification tokens‚Äù showing patterns of extracting specific recipe attributes.

---

# **5. How to Use These Insights in Continuous Memory Self-Evolution**

Here are actionable mechanisms:

---

## **Mechanism 1 ‚Äî Contrastive Memory Training**
- Positive prototypes = successful trajectory embeddings  
- Negative prototypes = failure patterns  

Loss:  
- Pull agent state embeddings toward successful prototypes  
- Push away from failure prototypes

---

## **Mechanism 2 ‚Äî Failure Retrieval as Early Warning**
If current state ‚âà failure prototype:
- trigger internal correction embedding  
- modify next-action generation  
- increase reliance on successful memory retrieval  

---

## **Mechanism 3 ‚Äî Correction Embedding Pairs**
For each failure token:
- generate a paired ‚Äúcorrect next action token‚Äù  
- store both  

At inference:
- encountering failure embedding ‚Üí retrieve correction embedding  

---

## **Mechanism 4 ‚Äî Workflow Guidance Memory**
Store successful micro-workflows:
- search ‚Üí select ‚Üí verify  
- filter ‚Üí apply ‚Üí select  

During inference:
- embed workflow priors into model‚Äôs next-step reasoning.

---

## **Mechanism 5 ‚Äî Popup-Recovery Memory**
Failure: popup leads to context reset  
Solution: store ‚Äúcorrect popup dismissal + return to task‚Äù trajectories.
